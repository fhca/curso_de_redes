```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.erdos_renyi_graph(20, .2)
nx.draw_networkx(G1)
plt.show()
```

```python
G2 = nx.barabasi_albert_graph(50, 1)
nx.draw_networkx(G2)
plt.show()
```

```python
pos = nx.spring_layout(G2, iterations=40)
nx.draw_networkx(G2, pos, node_size=60, with_labels=False)
plt.show()
```

```python
G3 = nx.grid_2d_graph(5,7)
pos = nx.spectral_layout(G3)
nx.draw_networkx(G3, pos, node_size=40, with_labels=False)
plt.show()
```

```python
def histo_grados(G1):
    n = G1.number_of_nodes()
    hist = nx.degree_histogram(G1)
    plt.bar(range(len(hist)), hist)
    print(hist)
    plt.show()

histo_grados(G1)
```
